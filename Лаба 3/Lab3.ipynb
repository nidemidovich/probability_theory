## Формула Бернулли

```python
import math
```

```python
def p_k_eq_m(m, n, p):
    q = 1 - p
    return math.comb(n, m) * pow(p, m) * pow(q, n - m)
```

```python
def p_k_less_m(m, n, p):
    return sum([p_k_eq_m(i, n, p) for i in range(m)])
```

```python
def p_k_greater_or_eq_m(m, n, p):
    return sum([p_k_eq_m(i, n, p) for i in range(m, n + 1)])
```

```python
def p_k_between(m_1, m_2, n, p):
    return sum([p_k_eq_m(i, n, p) for i in range(m_1, m_2 + 1)])
```

### Программа

$P_{n}(k=m) = C_{n}^{m}p^{m} \cdot q^{n-m}$

$P_{n}(k < m) = \sum_{i=0}^{m} P_{n}(k=i) = \sum_{i=0}^{m} C_{n}^{i}p^{i} \cdot q^{n-i}$

$P_{n}(k \geq m) = \sum_{i=m}^{n} P_{n}(k=i) = \sum_{i=m}^{n} C_{n}^{i}p^{i} \cdot q^{n-i}$

$P_{n}(m_{1} \leq k \leq m_{2}) = \sum_{i=m_{1}}^{m_{2}} P_{n}(k=i) = \sum_{i=m_{1}}^{m_{2}} C_{n}^{i}p^{i} \cdot q^{n-i}$

```python
print("""
    0. Выйти
    1. P(k=m)
    2. P(k<m)
    3. P(k>=m)
    4. P(m1<=m<=m2)""")
```

```python
choice = input('Что сделать? ')
while choice != '0':
    if choice == '1':
        m = int(input('Сколько раз должно произойти событие? '))
        n = int(input('Сколько испытаний проводится? '))
        p = float(input('Вероятность появления события = '))
        print(f'P(k=m) = {p_k_eq_m(m, n, p)}')
    
    elif choice == '2':
        m = int(input('До скольки раз должно произойти событие? '))
        n = int(input('Сколько испытаний проводится? '))
        p = float(input('Вероятность появления события = '))
        print(f'P(k<m) = {p_k_less_m(m, n, p)}')
        
    elif choice == '3':
        m = int(input('От скольки раз должно произойти событие? '))
        n = int(input('Сколько испытаний проводится? '))
        p = float(input('Вероятность появления события = '))
        print(f'P(k>=m) = {p_k_greater_or_eq_m(m, n, p)}')
        
    elif choice == '4':
        m_1 = int(input('От скольки раз должно произойти событие? '))
        m_2 = int(input('Какое максимальное количество раз должно произойти событие? '))
        n = int(input('Сколько испытаний проводится? '))
        p = float(input('Вероятность появления события = '))
        print(f'P(m1<=m<=m2) = {p_k_between(m_1, m_2, n, p)}')
    
    choice = input('Что сделать? ')
```

## Полиномилаьная формула

$P_{n}(m_{1}, m_{2}, ..., m_{k}) = \frac{n!}{m_{1}! m_{2}! ...m_{k}!}p_{1}^{m_{1}}\cdot p_{2}^{m_{2}} \cdot ... \cdot p_{k}^{m_{k}}$

```python
def polynom(n, m_list, p_list):
    denominator = 1
    for m in m_list:
        denominator *= math.factorial(m)
    
    p_product = 1
    for i in range(len(p_list)):
        p_product *= pow(p_list[i], m_list[i])
        
    return math.factorial(n) / denominator * p_product
```

### Программа

```python
n = int(input('Сколько независимых опытов проводится? '))
k = int(input('Сколько событий может произойти? '))

m_list = []
p_list = []
for i in range(k):
    m = int(input(f'm{i+1} = '))
    p = float(input(f'p{i+1} = '))
    m_list.append(m)
    p_list.append(p)
    
print(f'Искомая вероятность: {polynom(n, m_list, p_list)}')
```

## Формула Пуассона

```python
def poisson(m, n, p):
    lam = n * p
    return pow(lam, m) / math.factorial(m) * math.exp(-lam)
```

$P_{n}(X=m) \approx \frac{\lambda^{m}}{m!}e^{-\lambda},\ где\ \lambda = np$

```python
p = float(input('p (p>>0.9 || p<<0.1) = '))
n = int(input('n (n - велико) = '))
m = int(input('m = '))
print(f'Искомая вероятность: {poisson(m, n, p)}')
```
