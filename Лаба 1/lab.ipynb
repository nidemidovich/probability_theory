## Сочетания без повторений

$ C^{m}_{n}=\frac{n!}{(n-m)!} $

```python
import math
n = int(input('Количество элементов: '))
m = int(input('Выборка: '))
print(f'Количество сочетание из {m} по {n}: {math.comb(n, m)}')
```

## Перестановки с повторениями

$ P_{n}(m_{1}, m_{2}, ..., m_{k}) = \frac{n!}{m_{1}! \cdot m_{2}! \cdot ... \cdot m_{k}!}$

```python
def perm_with_rep(n, n_list):
    denominator = 1
    for i in n_list:
        denominator *= math.factorial(i)
        
    return math.factorial(n) / denominator
```

```python
n_list = []
n_i = None
i = 1
while n_i != '':
    try:
        n_i = int(input(f'n{i} = '))
    except:
        break
    n_list.append(n_i)
    i += 1
n = sum(n_list)
print(f'Количество перестановок с возвращениями P{n} = ', perm_with_rep(n, n_list))
```

## Задача 1

Имеется m операторов и n перенумерованных приборов, которые они могут обслуживать. Каждый оператор выбирает случайным образом и с одинаковой вероятностью любой прибор, но с условием, что ни один прибор не может обслуживаться больше, чем одним оператором. Найти вероятность того, что будут выбраны для обслуживания приборы с номерами 1, 2, ..., m.

$ P(A) = \frac{m!}{A^{m}_{n}} $

```python
def p1(m, n):
    A = math.factorial(n) / math.factorial(n - m)
    return math.factorial(m) / A
```

#### Тестовый пример

```python
n = int(input('Общее количество приборов: '))
m = int(input('Количество операторов: '))
print(f'Ответ: {p1(m, n)}')
```

## Задача 5

В ящике имеется k ТЭЗ, из них $ k_{1} $ элементов 1-го типа, ..., $ k_{i} $ элементов i-го типа, ..., $ k_{m} $ элементов m-го типа; $ \sum_{i=1}^{m} k_{i}=k $. Из ящика выбирают наугад n ТЭЗ. Найти вероятность того, что среди них будет $ n_{1} $ ТЭЗ 1-го типа, ..., $ n_{i} $ ТЭЗ i-го типа, ..., $ n_{m} $ ТЭЗ m-го типа.

$ n_{1} $ элементов из $ k_{1} $ элементов можно выбрать $ C_{k_{1}}^{n_{1}} $ способами. Аналогично, $ n_{m} $ элементов из $ k_{m} $ элементов можно выбрать $ C_{k_{m}}^{n_{m}} $ способами. Всего способов выбрать $ n $ элементов из ящика с $ k $ элементами $ C_{k}^{n} $.

Тогда вероятность будет равна
$$ P(A) = \frac{C_{k_{1}}^{n_{1}}...C_{k_{m}}^{n_{m}}}{C_{k}^{n}} = \frac{\prod_{i=1}^{m}C_{k_{i}}^{n_{i}}}{C_{k}^{n}} $$

```python
def p2(n_list, k_list):
    numerator = 1
    for i in range(len(n_list)):
        numerator *= math.comb(k_list[i], n_list[i])
    return numerator / math.comb(sum(k_list), sum(n_list))
```

#### Тестовый пример

```python
k = int(input('Количество типов ТЭЗ в ящике: '))
k_list = []
for i in range(k):
    k_i = int(input(f'k{i+1} = '))
    k_list.append(k_i)

print('Сколько ТЭЗ каждого типа должно быть? ')
n_list = []
for i in range(k):
    n_i = int(input(f'n{i+1} (n{i+1} < {k_list[i]}) = '))
    n_list.append(n_i)
    
print(f'Ответ: {p2(n_list, k_list)}')
```
