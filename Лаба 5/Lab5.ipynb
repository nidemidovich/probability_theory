```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

$\overline{x}_{в} = \frac{1}{n} \sum_{i=1}^{k} x_{i} \cdot n_{i}$

$D_{в} = \frac{1}{n} \sum_{i=1}^{k} (x_{i} - \overline{x}_{в})^2 \cdot n_{i}$

$\sigma_{в} = \sqrt{D_{в}}$

$S^2 = \frac{n}{n-1} D_{в}$

$S = \sqrt{S^2}$

# Задание 1

```python
np.random.seed(17)
rv = np.sort(np.random.randint(50, 80, 100))
```

```python
print('Вариационный ряд:')
print(rv)
```

```python
df = pd.DataFrame(index=range(len(np.unique(rv))))
df['Значение'] = pd.Series(np.unique(rv))
df['Частота'] = pd.Series(rv).value_counts().sort_index().to_list()
df['Частость'] = pd.Series(rv).value_counts(normalize=True).sort_index().to_list()
df.T
```

```python
plt.plot(df['Значение'], df['Частота'])
plt.title('Частота');
```

```python
plt.plot(df['Значение'], df['Частость'])
plt.title('Относительная частота');
```

```python
p = df['Частость'].values
```

```python
f = np.zeros(len(p)+1)
for i in range(1, len(f)):
    f[i] = np.sum(p[:i])
    
f
```

```python
plt.bar(np.unique(rv), f[1:], width=1, color='w', edgecolor='black', linewidth=1)
```

```python
mean = df.mean()['Частота']
variance = df.var()['Частота']
std = np.sqrt(variance)
fixed_var = len(rv) / (len(rv) - 1) * variance

print(f'Среднее: {mean}')
print(f'Дисперсия: {variance}')
print(f'СКО: {std}')
print(f'Исправленная дисперсия: {fixed_var}')
print(f'Исправленное СКО: {np.sqrt(fixed_var)}')
```

```python
intervals = np.array([44, 46, 48, 50, 52, 54, 56])
n = np.array([4, 13, 34, 32, 12, 5])
values = np.zeros(len(intervals)-1)
freq = np.zeros(len(intervals)-1)
for i in range(len(intervals)-1):
    values[i] = (intervals[i] + intervals[i+1]) / 2
    freq[i] = n[i] / np.sum(n)
print(values)
print(n)
print(freq)    
```

```python
rv = []
for i in range(len(n)):
    rv.extend([values.tolist()[i]]*n.tolist()[i])
rv = np.sort(np.array(rv))
print('Вариационный ряд: ')
print(rv)
```
