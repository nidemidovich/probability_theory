# Задание 1

Электрическая цепь составлена по схеме, приведенной на рисунке. События $ A_{i} $, $i = 1, 2, 3, 4, 5$ состоят в том, что одноименные элементы работают безотказно в течение времени $Т, P(A_{i}) = p_{i}$ . Найти вероятность события $В$ – схема работает безотказно в течение времени $Т$.

<img src="Task1.png">

Для безотказной работы схемы нужно, чтобы безотказно работали элементы $A_{1}, A_{2,3,4}$ и $A_{5}$.

Тогда событие $B$ будет представлено следующим образом: $B = A_{1} \cdot A_{2,3,4} \cdot A_{5}$, т.к. элементы соединены последовательно. 

Но элемент $A_{2,3,4}$ - это система из элементов $A_{2}, A_{3}, A_{4}$, соединенных параллельно. Чтобы ток точно прошел через эту систему, нужно, чтобы работал безотказно хотя бы один из элементов. Тогда событие $A_{2,3,4}$ можно представить в виде $$A_{2,3,4}=A_{2}+A_{3}+A_{4}$$

Итоговое событие: $$B = A_{1} \cdot (A_{2}+A_{3}+A_{4}) \cdot A_{5}$$

Вероятность наступления события $B$: $$P(B) = P(A_{1} \cdot (A_{2}+A_{3}+A_{4}) \cdot A_{5})$$

В силу независимости событий: 
$$P(B) = P(A_{1}) \cdot P(A_{2}+A_{3}+A_{4}) \cdot P(A_{5})$$
или, применяя теорему о вероятности хотя бы одного
$$P(B) = P(A_{1}) \cdot (1-P(\overline{A_{2}}) \cdot P(\overline{A_{3}}) \cdot P(\overline{A_{4}})) \cdot P(A_{5})$$

```python
def p1(p_a_list):
    p_a1, p_a2, p_a3, p_a4, p_a5 = p_a_list
    return p_a1 * (1 - (1 - p_a2) * (1 - p_a3) * (1 - p_a4)) * p_a5
```

```python
examples = [[0.6, 0, 0.7, 0, 0.8],
            [0.8, 0.6, 0, 0.7, 0.9],
            [0.5, 0.4, 0.6, 0.8, 0.7],
            [0.6, 0.8, 0.7, 0.9, 0.6]]

for l in examples:
    print(p1(l))
```

# Задание 3

Один студент выучил $m_{1}$ из $n$ вопросов программы, а второй $m_{2}$. Каждому из них задают по три вопроса. Найти  вероятность того, что на все три вопроса правильно ответят:


a) оба студента;\
b) только первый студент;\
c) только один студент;\
d) хотя бы один из студентов.

Для каждого случая распишем события:

$$A = A_{1}A_{2}$$
$$B = A_{1}\overline{A_{2}}$$
$$C = A_{1}\overline{A_{2}} + \overline{A_{1}}A_{2}$$
$$D = A_{1} + A_{2}$$

Благоприятный случай для первого студента - вытащить 3 билета из $m_{1}$ выученных; для второго - вытащить три билета из $m_{2}$ выученных. В общем случае они могут вытащить 3 из $n$ билетов.

Тогда
$$P(A_{1}) = \frac{C_{m_{1}}^{3}}{C_{n}^{3}}$$

$$P(A_{2}) = \frac{C_{m_{2}}^{3}}{C_{n}^{3}}$$

$$P(A) = P(A_{1}A_{2}) = P(A_{1})P(A_{2}) = \frac{C_{m_{1}}^{3}}{C_{n}^{3}} \frac{C_{m_{2}}^{3}}{C_{n}^{3}}$$

$$P(B) = P(A_{1}\overline{A_{2}}) = P(A_{1})P(\overline{A_{2}}) = P(A_{1})(1 - P(A_{2})) = \frac{C_{m_{1}}^{3}}{C_{n}^{3}}\Bigg(1 - \frac{C_{m_{2}}^{3}}{C_{n}^{3}}\Bigg)$$

$$P(C) = P(A_{1}\overline{A_{2}} + \overline{A_{1}}A_{2}) = P(A_{1}\overline{A_{2}}) + P(\overline{A_{1}}A_{2}) = \\ P(A_{1})(1 - P(A_{2})) + (1 - P(A_{1}))P(A_{2}) = \\ \frac{C_{m_{1}}^{3}}{C_{n}^{3}}\Bigg(1 - \frac{C_{m_{2}}^{3}}{C_{n}^{3}}\Bigg) + \Bigg(1 - \frac{C_{m_{1}}^{3}}{C_{n}^{3}}\Bigg)\frac{C_{m_{2}}^{3}}{C_{n}^{3}}$$

$$P(D) = P(A_{1} + A_{2}) = 1 - P(\overline{A_{1}})P(\overline{A_{2}}) = 1 -\Bigg(1 - \frac{C_{m_{1}}^{3}}{C_{n}^{3}}\Bigg)\Bigg(1 - \frac{C_{m_{2}}^{3}}{C_{n}^{3}}\Bigg)$$

```python
import math
```

```python
def p3(m1, m2, n):
    p1 = math.comb(m1, 3) / math.comb(n, 3)
    p2 = math.comb(m2, 3) / math.comb(n, 3)
    p_a = p1 * p2
    p_b = p1 * (1 - p2)
    p_c = p1 * (1 - p2) + (1 - p1) * p2
    p_d = 1 - (1 - p1) * (1 - p2)
    return [(p_a, 'A'), (p_b, 'B'), (p_c, 'C'), (p_d, 'D')]
```

## Пример

Один студент выучил 20 из 25 вопросов программы, а второй — только 15. Каждому из них задают по три вопроса. Найти вероятность того, что правильно ответят:

а) оба студента; \
б) только первый; \
в) только один из них; \
г) хотя бы один из студентов

$$P(A_{1}) = \frac{C_{20}^{3}}{C_{25}^{3}}$$

$$P(A_{2}) = \frac{C_{15}^{3}}{C_{25}^{3}}$$

$$P(A) = \frac{C_{20}^{3}}{C_{25}^{3}} \frac{C_{15}^{3}}{C_{25}^{3}} = 0.09805293005671077$$

$$P(B) = \frac{C_{20}^{3}}{C_{25}^{3}}\Bigg(1 - \frac{C_{15}^{3}}{C_{25}^{3}}\Bigg) = 0.3975992438563327$$

$$P(C) = \frac{C_{20}^{3}}{C_{25}^{3}}\Bigg(1 - \frac{C_{15}^{3}}{C_{25}^{3}}\Bigg) + \Bigg(1 - \frac{C_{20}^{3}}{C_{25}^{3}}\Bigg)\frac{C_{15}^{3}}{C_{25}^{3}} = 0.4973724007561437$$

$$P(D) = 1 -\Bigg(1 - \frac{C_{20}^{3}}{C_{25}^{3}}\Bigg)\Bigg(1 - \frac{C_{15}^{3}}{C_{25}^{3}}\Bigg) = 0.5954253308128544$$

```python
for p, event in p3(20, 15, 25):
    print(f'P({event}) = {p}')
```

# Задание 4

```python
def p_full(p_h: list, p_cond:list):
    return sum([p_h[i] * p_cond[i] for i in range(len(p_h))])
```

```python
def bayes(num: int, p_h: list, p_cond: list):
    i = num - 1
    return p_h[i] * p_cond[i] / p_full(p_h, p_cond)
```

## Задача на полную вероятность

В группе спортсменов лыжников в 2 раза больше, чем бегунов, а бегунов в 3 раза больше, чем велосипедистов. Вероятность выполнить норму для лыжника 0,9, для бегуна 0,75, для велосипедиста - 0,8. Найти вероятность того, что спортсмен, выбранный наугад, выполнит норму.

$H_{1} = \{Спортсмен - лыжник\} \\
H_{2} = \{Спортсмен - бегун\} \\
H_{3} = \{Спортсмен - велосипедист\}$

Велосипедистов $x$, бегунов $3x$, лыжников $6x$. Найдем вероятности гипотез:

$ P(H_{1}) = \frac{6x}{x+3x+6x} = \frac{6}{10} = 0.6 \\
P(H_{1}) = \frac{3x}{x+3x+6x} = \frac{3}{10} = 0.3 \\
P(H_{1}) = \frac{x}{x+3x+6x} = \frac{1}{10} = 0.1$

Событие $A$ - спортсмен выполнит норму. Известны условные вероятности:

$P(A | H_{1}) = 0.9 \\
P(A | H_{2}) = 0.75 \\
P(A | H_{3}) = 0.8$

И тогда полная вероятность:

$P(A) = P(A | H_{1})P(H_{1}) + P(A | H_{2})P(H_{2}) + P(A | H_{3})P(H_{3}) = 0.6 \cdot 0.9 + 0.3 \cdot 0.75 + 0.1 \cdot 0.8 = 0.845$

```python
p_h = [0.6, 0.3, 0.1]
p_cond = [0.9, 0.75, 0.8]
```

```python
print(f'P(A) = {p_full(p_h, p_cond)}')
```

```python
h_num = int(input('Количество гипотез: '))
p_h = []
for i in range(h_num):
    p = float(input(f'P(H{i+1}) = '))
    p_h.append(p)
  
p_cond = []
for i in range(h_num):
    p = float(input(f'P(A|H{i+1}) = '))
    p_cond.append(p)
    
print(f'P(A) = {p_full(p_h, p_cond)}')
```
